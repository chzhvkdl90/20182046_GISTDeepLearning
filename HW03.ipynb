## Guide2. How to use the DataLoader
***


- [Torchvision](https://pytorch.org/docs/stable/torchvision/index.html) is a package in the PyTorch library containing computer-vision models, datasets, and image transformations. 


```
import torch # import PyTorch Library
import torchvision ## import TorchVision package
import torch.nn as nn
import torchvision.transforms as transforms

train_dir = 'train' # path of train folder 
test_dir = 'test' # path of test folder
```

- Define some hyperparameters

```
# Hyper parameters
num_epochs = 20 # number of epochs for train
batch_size = 100 # do not change this value
learning_rate = 0.0001 # do not change this value
```

- Define data loader to train and test the model

```
# Image transformation
transform = transforms.Compose([ # Composes several transforms together.
    transforms.Grayscale(), # Convert image to grayscale.
    transforms.ToTensor() # Convert a PIL Image to tensor.
])

# Image folder
train_dataset = torchvision.datasets.ImageFolder(root=train_dir, transform=transform)
test_dataset = torchvision.datasets.ImageFolder(root=test_dir, transform=transform)

# Data loader
# In train step, we should shuffle the order of input sequence to prevent overfitting.
train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)
test_loader = torch.utils.data.DataLoader(dataset=test_dataset, batch_size=batch_size, shuffle=False)
```

## Problem 2. Visualization (60%)
___

- First, Let's instantiate our model and define criterion and optimizer to train our model.

```
model = LogisticRegression()
print(model)
# Loss and optimizer

criterion = torch.nn.BCELoss(reduction='mean')
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

# It will print an error if your model has not been built correctly.
```

- This is a training loop for our logistic regression model.  
By using tensorboard X,  
- **Problem 2-1** visualize the loss function (x-axis: epoch or step_size, y-axis: loss)  
- **Problem 2-2** visualize the test accuracy (x-axis: epoch, y-axis: test accuracy)  

By using matplotlib,  
- **Problem 2-3** visualize the loss function (x-axis: epoch or step_size, y-axis: loss)  
- **Problem 2-4** visualize the test accuracy (x-axis: epoch, y-axis: test accuracy)  
**Note: Upload your graph image on GEL along with .ipynb **

```
for epoch in range(num_epochs):
    for i, (images, labels) in enumerate(train_loader): # Get the train batch
        model.train() # Sets the module in train mode
        labels = labels.type(torch.FloatTensor) # change DoubleTensor to FloatTensor

        ### 1. Forward pass
        outputs = model(images) # pass the images to model
        loss = criterion(outputs, labels) # compute the loss function

        ### 2. Backward pass and optimization
        # Zero gradients, perform a backward pass, and update the weights.
        optimizer.zero_grad() # Clears the gradients of all optimized
        loss.backward() # Compute the gradients with respect to loss
        optimizer.step() # Performs a single optimization step (parameter update)
        
    ### 3. Evalaluate the model in test dataset
    model.eval()  # Sets the module in evaluation mode
    # compute test accuracy
    with torch.no_grad():
        correct = 0
        total = 0
        for images, labels in test_loader: # Get the test batch
            labels = labels.type(torch.FloatTensor)
            outputs = model(images).data # pass the test image to model
            predicted = torch.round(outputs) # round the outputs (if outputs > 0.5, output == 1. else, outputs == 0)
            total += labels.size(0)        
            correct += (predicted == labels).sum().item()
        test_acc = 100 * (correct / total)
        print('Epoch', epoch, ': Accuracy on the 10000 test images', test_acc)           
```
